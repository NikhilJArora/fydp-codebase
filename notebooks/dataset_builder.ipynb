```python
# consts
# image folder:
r_images_dir = '/Users/nikhilarora/data/fydp/sb_images/orig'


# output location:
data_car_dir = '/Users/nikhilarora/data/fydp/dataset/car'
data_ncar_dir = '/Users/nikhilarora/data/fydp/dataset/ncar'
```

```python
# grab a photo
# ADD PHOTO NAME WITHIN WORKING DIR:
photo_name = 'IMG_9842.JPG' 
r_img = np.flip(np.flip(io.imread(os.path.join(r_images_dir, photo_name)),1),0)
```

```python
# look at the photo in a grid 
plt.imshow(r_img,cmap='gray')
```

```python
# define the starting and ending values for slicing each spot 
# (ystart, yend, xstart, xend)
spots = [
    (1600, 2300, 500, 1700),
    (1600, 2300, 2000, 3400),
    #(2000, 2500, 2600, 4000)
]
```
